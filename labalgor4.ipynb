```python
# Александрова Екатерина
# Группа: ЦИБ-251
#Напишите функцию get_md5_hash(s), которая принимает строку s и
#возвращает её MD5-хеш в шестнадцатеричном (hex) формате.

import hashlib

def get_md5_hash(s: str) -> str:
    """
    Функция get_md5_hash(s) принимает строку, кодирует её в UTF-8, вычисляет MD5-хеш
    с помощью библиотеки hashlib и возвращает результат в шестнадцатеричном формате.
    s: входная строка
    return: MD5-хеш в hex-формате
    """
    md5 = hashlib.md5()          # создаём объект MD5
    md5.update(s.encode('utf-8'))  # кодируем строку в байты
    return md5.hexdigest()       # возвращаем хеш


if __name__ == "__main__":
    # Тест 1
    print(get_md5_hash("Hello"))
    # Ожидаемый результат:
    # 8b1a9953c4611296a827abf8c47804d7

    # Тест 2
    print(get_md5_hash("admin"))
    # Ожидаемый результат:
    # 21232f297a57a5a743894a0e4a801fc3
```
